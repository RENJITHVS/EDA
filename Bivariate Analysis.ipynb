```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

# 1.ScatterPlot (Numerical - Numerical)

```python
sns.scatterplot(data.Fare,data.Age,hue=data["Sex"],size=data.Embarked)
```

## 2. Barplot

```python
sns.barplot(data.Pclass,data.Age,hue=data.Sex)
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data.Sex, data.Age, hue= data.Survived)
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(data[data.Survived == 0]['Age'])
sns.distplot(data[data.Survived == 1]['Age'])
```

## 5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(data['Pclass'],data['Survived']))
```

```python
(data.groupby('Embarked').mean()['Survived']*100)
```

```python
(data.groupby('Embarked').mean()['Survived']*100).plot(kind='bar')
```

```python
data.groupby("Sex").mean()['Survived'] * 100
```

```python
(data.groupby("Sex").mean()['Survived'] * 100).plot(kind='pie')
```

## 6. ClusterMap (Categorical - Categorical)

```python
pd.crosstab(data['Parch'],data['Survived'])
```

```python
sns.clustermap(pd.crosstab(data['Parch'],data['Survived']))
```
